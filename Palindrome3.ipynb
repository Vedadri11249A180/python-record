```python
Q3="""Write a Python program that takes a number as input, determines whether it is a palindrome, and prints the number of times each digit from 0 to 9 appears in the number"""
print(Q3)

number = input("Enter a number: ")
if number == number[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")

for i in range(10):
    count = number.count(str(i))
    if count > 0:
        print(f"{i} appears {count} times")
```
